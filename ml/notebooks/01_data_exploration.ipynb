# Data Exploration

In this notebook, we will analyze our dataset to check the class distribution. This is a crucial step before training our PyTorch model.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../../data/diabetes.csv")
df.head()
```

```python
# Check class balance (counts)
print("Class distribution (counts):")
print(df["Outcome"].value_counts())

# Check class balance (percentages)
print("\nClass distribution (percentages):")
print(df["Outcome"].value_counts(normalize=True) * 100)
```

```python
# Visualization (Bar Chart)
sns.countplot(data=df, x="Outcome", palette="coolwarm")
plt.title("Class Balance in Diabetes Dataset (0 = Healthy, 1 = Diabetic)")
plt.show()
```

```python
# Visualization (Pie Chart with percentages)
plt.figure(figsize=(6, 6))
df["Outcome"].value_counts().plot.pie(
    autopct="%1.1f%%",
    labels=["0 (Healthy)", "1 (Diabetic)"],
    colors=["#66b3ff", "#ff9999"],
    startangle=90,
    shadow=True,
)
plt.title("Class Balance (Percentage)")
plt.ylabel("")
plt.show()
```

### Analysis Conclusions:
- As we can see, class 0 (healthy) dominates the dataset (approx. 65%).
- When we split this data into training (Train), validation (Val), and test (Test) sets, we must use a **stratified split**.
- This will ensure that the exact same proportion (65/35) is maintained in each subset, allowing our neural network to learn representative features of both groups.
